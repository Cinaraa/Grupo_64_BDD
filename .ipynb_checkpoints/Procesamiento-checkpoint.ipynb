Función que limpia carácteres especiales

```python
def limpiar_especiales(fila):
    for atributo in fila:
        i = 0
        for caracter in atributo:
            if caracter == "Ñ":
                caracter[i] = "N"
            elif caracter == "ñ":
                caracter[i] = "n"
            elif caracter == "Á":
                caracter[i] = "A"
            elif caracter == "á":
                caracter[i] = "a"
            elif caracter == "É":
                caracter[i] = "E"
            elif caracter == "é":
                caracter[i] = "e"
            elif caracter == "Í":
                caracter[i] = "I"
            elif caracter == "í":
                caracter[i] = "i"
            elif caracter == "ñ":
                caracter[i] = "n"
            
```
