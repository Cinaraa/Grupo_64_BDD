```python
import pandas as pd
import numpy as np
```

```python
artistas = pd.read_csv('Datos Pares/Artistas.csv')
eventos = pd.read_csv('Datos Pares/Evento.csv')
productoras = pd.read_csv('Datos Pares/Productoras.csv')
recintos = pd.read_csv('Datos Pares/Recinto.csv')
entradas = pd.read_csv('Datos Pares/Entradas.csv')
```

```python
artistas.rename(columns = {'Nombre artistico':'nombre_artistico', 'Nº contacto':'contacto_artista'}, inplace = True)
entradas.rename(columns = {'Evento':'nombre_evento', 'numero asiento':'numero_asiento'}, inplace = True)
recintos.rename(columns = {'Nombre':'nombre_recinto', 'Ciudad':'ciudad', 'Pais':'pais', 'Capacidad':'capacidad'}, inplace = True)
productoras.rename(columns = {'Nombre':'nombre_productora', 'País':'pais', 'Inicio actividades':'inicio_actividades', 'Nº contacto':'contacto_productora'}, inplace = True)
eventos.rename(columns = {'Nombre':'nombre_evento', 'Recinto':'recinto', 'Fecha de incio':'fecha_inicio', 'Fecha de termino':'fecha_termino', 'Artista':'artista', 'Productora':'productora'}, inplace = True)
```

```python
artistas.drop_duplicates(inplace=True, ignore_index=True)
eventos.drop_duplicates(inplace=True, ignore_index=True)
productoras.drop_duplicates(inplace=True, ignore_index=True)
recintos.drop_duplicates(inplace=True, ignore_index=True)
entradas.drop_duplicates(inplace=True, ignore_index=True)
```

```python
productoras.drop_duplicates(subset=['nombre_productora'], inplace=True, ignore_index=True)
recintos.drop_duplicates(subset=['nombre_recinto'], inplace=True, ignore_index=True)
```

```python
artistas['nombre_artistico'] = artistas['nombre_artistico'].str.strip()

eventos['nombre_evento'] = eventos['nombre_evento'].str.strip()
eventos['recinto'] = eventos['recinto'].str.strip()
eventos['artista'] = eventos['artista'].str.strip()
eventos['productora'] = eventos['productora'].str.strip()

productoras['nombre_productora'] = productoras['nombre_productora'].str.strip()
productoras['pais'] = productoras['pais'].str.strip()

recintos['nombre_recinto'] = recintos['nombre_recinto'].str.strip()
recintos['ciudad'] = recintos['ciudad'].str.strip()
recintos['pais'] = recintos['pais'].str.strip()

entradas['nombre_evento'] = entradas['nombre_evento'].str.strip()
entradas['categoria'] = entradas['categoria'].str.strip()
entradas['numero_asiento'] = entradas['numero_asiento'].str.strip()
```

```python
eventos.index.name = 'id_evento'
artistas.index.name = 'id_artista'
productoras.index.name = 'id_productora'
recintos.index.name = 'id_recinto'
entradas.index.name = 'id_entrada'
```

```python
eventos["fecha_inicio"] = eventos["fecha_inicio"].str.replace("-","/")
eventos["fecha_termino"] = eventos["fecha_termino"].str.replace("-","/")
productoras["inicio_actividades"] = productoras["inicio_actividades"].str.replace("-","/")
```

```python
eventos.fecha_inicio = eventos.fecha_inicio.replace({'30/2/':'28/02/'}, regex=True)
eventos.fecha_inicio = eventos.fecha_inicio.replace({'30/02/':'28/02/'}, regex=True)

eventos.fecha_inicio = eventos.fecha_inicio.replace({'31/04/':'30/04/'}, regex=True)
eventos.fecha_inicio = eventos.fecha_inicio.replace({'31/4/':'30/04/'}, regex=True)

eventos.fecha_termino = eventos.fecha_termino.replace({'30/02/':'28/02/'}, regex=True)
eventos.fecha_termino = eventos.fecha_termino.replace({'30/2/':'28/02/'}, regex=True)

eventos.fecha_termino = eventos.fecha_termino.replace({'31/04/':'30/04/'}, regex=True)
eventos.fecha_termino = eventos.fecha_termino.replace({'31/4/':'30/04/'}, regex=True)

productoras.inicio_actividades = productoras.inicio_actividades.replace({'30/2/':'28/02/'}, regex=True)
productoras.inicio_actividades = productoras.inicio_actividades.replace({'30/02/':'28/02/'}, regex=True)
productoras.inicio_actividades = productoras.inicio_actividades.replace({'31/04/':'30/04/'}, regex=True)
productoras.productoras = productoras.inicio_actividades.replace({'31/4/':'30/04/'}, regex=True)
```

```python
eventos['fecha_inicio'] =pd.to_datetime(eventos.fecha_inicio)
eventos['fecha_termino'] =pd.to_datetime(eventos.fecha_termino)
productoras['inicio_actividades'] =pd.to_datetime(productoras.inicio_actividades)
```
